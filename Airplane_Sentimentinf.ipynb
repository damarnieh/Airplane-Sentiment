```python
import tensorflow as tf
import numpy as np
import pandas as pd
import pickle

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
model = load_model('best_ann_sentiment_model.keras')
model.summary()
```

```python
with open('tokenizer.pkl', 'rb') as f:
    tokenizer = pickle.load(f)
```

```python
label_map = {
    0: 'Negative',
    1: 'Neutral',
    2: 'Positive'
}
```

```python
def predict_sentiment(text):
    # tokenisasi
    seq = tokenizer.texts_to_sequences([text])
    pad = pad_sequences(seq, maxlen=50, padding='post')
    
    # prediksi
    pred = model.predict(pad)
    label_index = np.argmax(pred)
    
    return label_map[label_index]
```

```python
sample_text = "The flight was delayed and the service was terrible"

result = predict_sentiment(sample_text)
print("Tweet :", sample_text)
print("Sentiment Prediction :", result)
```

```python
sample_texts = [
    "I love flying with this airline",
    "The flight was okay, nothing special",
    "Worst experience ever, very disappointed"
]

for text in sample_texts:
    print(f"{text} --> {predict_sentiment(text)}")
```
